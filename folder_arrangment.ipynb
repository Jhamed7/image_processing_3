```python
import os
import re
```

### images should be in a folder in same path with this file

```python
path = os.getcwd()
print(path)
```

```python
os.listdir(path+'\\MNIST_persian')
```

```python
for num in range(10):
    os.makedirs(os.path.join(path, f'new_MNIST\\{num}'))
```

```python
destination = os.path.join(path + '\\new_MNIST')
for (root,dirs,files) in os.walk(path + '\\MNIST_persian', topdown=True):
    dest_num = re.sub('\D', '', root[-3:])
    for number in range(10):
        for file_ in files:
            if file_.endswith('jpg') and file_[0] == f'{number}':
                os.rename(os.path.join(root,file_), os.path.join(destination, f'{number}\\{dest_num+"_"+file_}'))
```
