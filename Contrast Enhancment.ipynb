```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Lenna.png', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('lenna_gray.jpg', image)
```

```python
rows, cols = image.shape
```

```python
rows, cols
```

```python
plt.imshow(image, cmap='gray')
```

```python
def my_func(x):
    y = (258 / (1 + np.exp(-0.039 * (x-130)))) - 1.5
    return y
```

```python
def Gamma_correction(x, gamma):
    y = round((x/255)**gamma * 255)
    return y
```

```python
def func_on_image(img, func, *args):
    rows, cols = img.shape
    out = np.copy(img)
    for i in range(rows):
        for j in range(cols):
            out[i, j] = func(img[i, j], *args)
    return out
```

# contrast enhancment with 2 methods

### 1- a custom nonlinear function

### 2- Gamma correction function in link : https://docs.opencv.org/3.4/d3/dc1/tutorial_basic_linear_transform.html


#### method 1

```python
enhanced_lenna = func_on_image(image, my_func)
cv2.imwrite('enhanced_lenna.jpg', enhanced_lenna)
```

```python
plt.imshow(enhanced_lenna, cmap='gray')
```

#### mehod 2

```python
g_enhanced_lenna = func_on_image(image, Gamma_correction, 3.5)  # Gamma(γ) = 3.5
```

```python
plt.imshow(g_enhanced_lenna, cmap='gray')
```
